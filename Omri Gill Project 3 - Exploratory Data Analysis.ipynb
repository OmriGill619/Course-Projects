<b>Student answer.</b> <a class="tocSkip"></a>
    Our first attempt at replacing NaN values across the dataset.
    I defined a function that receives the parameters "column_to_replace" (the column whose missing values we want to replace) and "reference_colummns" (the columns according to which we will replace the NaN values).
    For example: we want to replace the NaN values in the "cylinders" column, based upon similar values for rows with no missing values in the columns 'model', 'model_year', 'fuel'.
    
In the process, I found a method ".mode()", which return the most common value in a DataFrame column. This value would be the one assigned to replace the NaN.
    
I am fully aware that this might be a bit complicated (and, frankly, takes a while to run) - for example, I could have just used fillna() using the median - but I wanted to experiment, and to fill the NaN with the most appropriate values.
    
I used both a conditional dataset, and using query.

```python
def replace_missing(column_to_replace, reference_columns, default_value):
    data_with_missing = data[data[column_to_replace].isna()]
    for index, row in data_with_missing.iterrows():
        #query_string = 'model==@row[reference_columns[0]]'
        
        #print(query_string)
        var00 = data[reference_columns[0]]
        var01 = row[reference_columns[0]]
        var10 = data[reference_columns[1]]
        var11 = row[reference_columns[1]]
        var20 = data[reference_columns[2]]
        var21 = row[reference_columns[2]]

        reference_data = data.query('@var00==@var01 and @var10==@var11 and @var20==@var21')
#       reference_data = data[(data[reference_columns[0]] == row[reference_columns[0]]) &
#                             (data[reference_columns[1]] == row[reference_columns[1]]) &
#                             (data[reference_columns[2]] == row[reference_columns[2]])]
        if (reference_data[column_to_replace].mode().empty == True): #the reference data set is empty. use default
            x = default_value
        else:
            x=reference_data[column_to_replace].mode()[0] #most common value in column
        data.loc[index, column_to_replace] = x
    return 
   
replace_missing('cylinders', ['model', 'model_year', 'fuel'], data['cylinders'].median())
replace_missing('model_year', ['model', 'cylinders', 'fuel'], data['model_year'].median())

print(data['cylinders'].isna().sum())
print(data['model_year'].isna().sum())
```

    I've created a function to eliminate all rare cases (under 50). I've run the function twice for each type, once for paint_color and once for condition. I could have done these attributes in a single function, but wanted to see the intermediate process.

```python
def exclude_rare_cases(df, column):
    df_column_dict = df[column].value_counts().to_dict()
    remove_rare = []
    for attribute,value in df_column_dict.items():
        if value <50:
            remove_rare.append(attribute)   
    df = df[~df[column].isin(remove_rare)]
    return df

suv_data = exclude_rare_cases(suv_data, 'paint_color')
suv_data = exclude_rare_cases(suv_data, 'condition')
sedan_data = exclude_rare_cases(sedan_data, 'paint_color')
sedan_data = exclude_rare_cases(sedan_data, 'condition')
truck_data = exclude_rare_cases(truck_data, 'paint_color')
truck_data = exclude_rare_cases(truck_data, 'condition')




```
