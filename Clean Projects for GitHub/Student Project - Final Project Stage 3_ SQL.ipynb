Now we will handle our second task - find the number of user reviews and the average rating for each book.

```python
merge_query_q2_updated = """
SELECT books.book_id, books.title, COALESCE(num_reviews, 0) AS num_reviews, COALESCE(avg_rating, 0) AS avg_rating
FROM books
LEFT JOIN (
    SELECT book_id, COUNT(review_id) AS num_reviews
    FROM reviews
    GROUP BY book_id
) AS subquery_reviews ON books.book_id = subquery_reviews.book_id
LEFT JOIN (
    SELECT book_id, AVG(rating) AS avg_rating
    FROM ratings
    GROUP BY book_id
) AS subquery_ratings ON books.book_id = subquery_ratings.book_id
"""
merged_df_q2_updated = pd.read_sql_query(merge_query_q2_updated, engine)
merged_df_q2_updated
```

Now we will handle our third task - identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).

```python
query_q3 = """
    SELECT publishers.publisher_id, publishers.publisher, COUNT(books.book_id) AS book_count
    FROM books
    JOIN publishers ON books.publisher_id = publishers.publisher_id
    WHERE books.num_pages > 50
    GROUP BY publishers.publisher_id, publishers.publisher
    ORDER BY book_count DESC
    LIMIT 1
"""
result_q3 = pd.read_sql_query(query_q3, engine)
result_q3
```

Now we will handle our fourth task - identify the author with the highest average book rating (look only at books with at least 50 ratings).

```python
query_q4 = """
    SELECT authors.author_id, authors.author, AVG(ratings.rating) AS avg_rating
    FROM books
    JOIN authors ON books.author_id = authors.author_id
    JOIN ratings ON books.book_id = ratings.book_id
    WHERE books.book_id IN (
        SELECT book_id
        FROM ratings
        GROUP BY book_id
        HAVING COUNT(*) >= 50
    )
    GROUP BY authors.author_id, authors.author
    ORDER BY avg_rating DESC
    LIMIT 1
"""
result_q4 = pd.read_sql_query(query_q4, engine)
result_q4
```

Now we will handle our fifth task - find the average number of text reviews among users who rated more than 50 books.

```python
query_q5 = """
    SELECT AVG(review_count) AS avg_text_reviews
    FROM (
        SELECT r.username, COUNT(DISTINCT rv.review_id) AS review_count
        FROM ratings r
        JOIN reviews rv ON r.username = rv.username
        WHERE r.username IN (
            SELECT username
            FROM ratings
            GROUP BY username
            HAVING COUNT(*) > 50
        )
        GROUP BY r.username
    ) AS subquery
"""
result_q5 = pd.read_sql_query(query_q5, engine)
print("Average number of text reviews among users who rated more than 50 books:", result_q5['avg_text_reviews'].values[0])
```
